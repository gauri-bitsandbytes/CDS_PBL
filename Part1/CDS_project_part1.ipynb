### *Task 1*

Build a data loader for the validation dataset present in the following path: `data_students/student_dataset.hdf5`. You will be using this dataset to validate the performance of the ML model. The dataset is in HDF5 binary data format. This format is used to store large amount of data. Make sure that you import and familiarise yourself with the right Python libraries to handle HDF5 files.


```python
# h5py is used to open and read the HDF5 dataset file.
import h5py
# NumPy is used for array operations like counting labels and selecting random rows.
import numpy as np
# pandas is used to show results as readable tables in the notebook.
import pandas as pd
# torch is the main PyTorch library used for tensors and model execution.
import torch
# nn contains PyTorch building blocks for neural networks, such as Linear, ReLU, and Sigmoid.
from torch import nn
# TensorDataset pairs feature tensors with label tensors; DataLoader creates batches from them.
from torch.utils.data import DataLoader, TensorDataset

# Path to the validation dataset provided for the lab.
DATA_PATH = "student_dataset.hdf5"
# Path to the pretrained model file provided for the lab.
MODEL_PATH = "model_2023-03-28_20-03.pth"
# Number of samples processed at once during prediction.
BATCH_SIZE = 64

# Open the HDF5 file in read-only mode.
with h5py.File(DATA_PATH, "r") as file:
    # Load the numeric feature vectors used as model input.
    X = file["vectors"][:].astype(np.float32)
    # Load the binary labels: 1 means vulnerable, 0 means non-vulnerable.
    y = file["labels"][:].astype(np.float32)
    # Load the original source-code samples for display and inspection.
    code_samples = file["source"][:]

# Convert the feature array into a PyTorch tensor so it can be passed into the model.
X_tensor = torch.tensor(X, dtype=torch.float32)
# Convert the label array into a PyTorch tensor and reshape it to match the model output shape.
y_tensor = torch.tensor(y, dtype=torch.float32).view(-1, 1)

# Combine every feature vector with its corresponding label.
dataset = TensorDataset(X_tensor, y_tensor)
# Create batches from the dataset; shuffle is False because this is evaluation, not training.
validation_loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=False)

# Print the number of validation samples to confirm the dataset was loaded correctly.
print(f"Loaded {len(dataset)} validation samples.")
# Print the feature shape; here it should be 1000 samples with vectors of size 768.
print(f"Feature shape: {X.shape}")
# Print the label shape; here it should contain one label for each sample.
print(f"Label shape: {y.shape}")
```

### *Task 2*

Generate a table with 10 random samples from the dataset and show their corresponding labels.


```python
# Create a random generator with seed 42 so the selected samples are the same every run.
rng = np.random.default_rng(42)
# Choose 10 random sample indices from the dataset without repeating an index.
sample_indices = rng.choice(len(y), size=10, replace=False)

# This list will store one dictionary for each row in the final table.
rows = []
# Go through each randomly selected index.
for idx in sample_indices:
    # Read the source-code sample at this index.
    sample = code_samples[idx]

    # Some HDF5 text values are stored as bytes, so convert them to normal strings.
    if isinstance(sample, bytes):
        # Decode bytes using UTF-8; replace unknown characters instead of raising an error.
        sample = sample.decode("utf-8", errors="replace")

    # Add sample information to the table row list.
    rows.append({
        # Original index of the sample inside the validation dataset.
        "sample_id": int(idx),
        # Short preview of the source code; line breaks are replaced to keep the table compact.
        "sample_preview": str(sample).replace("\n", " ")[:250],
        # Numeric label from the dataset.
        "label": int(y[idx]),
        # Human-readable form of the numeric label.
        "label_name": "vulnerable" if y[idx] == 1 else "non-vulnerable",
    })

# Convert the collected rows into a pandas table and display it.
pd.DataFrame(rows)
```

### *Task 3*

Inspect the dataset and answer the following questions:
1. How many samples are in the dataset?
2. How many positive examples (vulnerability-labeled instances) are in the dataset?
3. What is the vulnerable/non-vulnerable ratio?


```python
# Total number of samples is the length of the label array.
num_samples = len(y)
# Count labels equal to 1, which represent vulnerable examples.
num_positive = int((y == 1).sum())
# Count labels equal to 0, which represent non-vulnerable examples.
num_negative = int((y == 0).sum())
# Calculate how many vulnerable samples exist compared with non-vulnerable samples.
vulnerable_ratio = num_positive / num_negative

# Store the dataset statistics in a table.
dataset_summary = pd.DataFrame({
    # Text descriptions of the values we calculated.
    "question": [
        "Number of samples",
        "Positive/vulnerable examples",
        "Negative/non-vulnerable examples",
        "Vulnerable / non-vulnerable ratio",
    ],
    # The actual calculated values.
    "answer": [
        num_samples,
        num_positive,
        num_negative,
        round(vulnerable_ratio, 4),
    ],
})

# Display the dataset summary table.
dataset_summary
```

### *Task 4*

Load and run the following pre-trained neural network model called `VulnPredictionModel`.


```python
# Check whether PyTorch can use an NVIDIA CUDA GPU.
device = "cuda" if torch.cuda.is_available() else "cpu"
# Print the selected device so we know where the model will run.
print(f"Using {device} device")
```

```python
# Define the given pretrained model architecture from the lab.
class VulnPredictionModel(nn.Module):
    # __init__ creates the layers of the neural network.
    def __init__(self):
        # Initialize the parent PyTorch Module class.
        super().__init__()
        # Flatten changes input shape from (batch, 1, 768) to (batch, 768).
        self.flatten = nn.Flatten()
        # linear_stack contains the same layers as the provided pretrained model.
        self.linear_stack = nn.Sequential(
            # First fully connected layer: 768 input features to 64 hidden values.
            nn.Linear(768, 64),
            # ReLU activation keeps positive values and sets negative values to zero.
            nn.ReLU(),
            # Second fully connected layer: 64 hidden values to 64 hidden values.
            nn.Linear(64, 64),
            # Second ReLU activation adds non-linearity again.
            nn.ReLU(),
            # Final layer: 64 hidden values to one output value.
            nn.Linear(64, 1),
            # Sigmoid converts the output to a probability between 0 and 1.
            nn.Sigmoid(),
        )

    # forward defines how input data moves through the model.
    def forward(self, x):
        # Flatten the input before passing it to the linear layers.
        x = self.flatten(x)
        # Return the final vulnerability probability.
        return self.linear_stack(x)


# Create the given model architecture and move it to CPU or GPU.
model = VulnPredictionModel().to(device)
# Load the pretrained weights from model_2023-03-28_20-03.pth into this model.
model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
# Print the model file name so it is clear which pretrained model is used.
print(f"Using pretrained model: {MODEL_PATH}")
# Put the model in evaluation mode because we are not training it.
model.eval()

# Display the loaded model architecture.
model
```

### *Task 5*

Make a prediction on the provided dataset and compute the following values:
- True Positives
- True Negatives
- False Positives
- False Negatives


```python
# Store model output probabilities for all batches.
all_predictions = []
# Store true labels for all batches.
all_labels = []

# Disable gradient calculation to make evaluation faster and use less memory.
with torch.no_grad():
    # Loop over the validation data batch by batch.
    for features, labels in validation_loader:
        # Move the feature batch to the same device as the model.
        features = features.to(device)
        # Run the model and move predictions back to CPU for metric calculation.
        predictions = model(features).cpu()

        # Save this batch's predictions.
        all_predictions.append(predictions)
        # Save this batch's true labels.
        all_labels.append(labels)

# Combine prediction batches into one flat tensor.
probabilities = torch.cat(all_predictions).view(-1)
# Combine label batches into one flat integer tensor.
true_labels = torch.cat(all_labels).view(-1).int()
# Convert probabilities to binary predictions using threshold 0.5.
predicted_labels = (probabilities >= 0.5).int()

# True positive: predicted vulnerable and actually vulnerable.
tp = int(((predicted_labels == 1) & (true_labels == 1)).sum())
# True negative: predicted non-vulnerable and actually non-vulnerable.
tn = int(((predicted_labels == 0) & (true_labels == 0)).sum())
# False positive: predicted vulnerable but actually non-vulnerable.
fp = int(((predicted_labels == 1) & (true_labels == 0)).sum())
# False negative: predicted non-vulnerable but actually vulnerable.
fn = int(((predicted_labels == 0) & (true_labels == 1)).sum())

# Put the confusion counts into a readable table.
confusion_counts = pd.DataFrame(
    {"count": [tp, tn, fp, fn]},
    index=["True Positives", "True Negatives", "False Positives", "False Negatives"],
)

# Display the confusion-count table.
confusion_counts
```

### *Task 6*

Compute the corresponding performance metrics **manually** (do not use PyTorch's predefined metrics):
- Accuracy
- Precision
- Recall
- F1


```python
# Accuracy: percentage of all predictions that are correct.
accuracy = (tp + tn) / (tp + tn + fp + fn)
# Precision: when the model predicts vulnerable, how often it is correct.
precision = tp / (tp + fp)
# Recall: how many actually vulnerable samples the model finds.
recall = tp / (tp + fn)
# F1-score: balance between precision and recall.
f1_score = 2 * precision * recall / (precision + recall)

# Store the metrics in a table.
metrics = pd.DataFrame({
    "metric": ["Accuracy", "Precision", "Recall", "F1-score"],
    "value": [accuracy, precision, recall, f1_score],
})

# Display the final metric table.
metrics
```

### *Task 7*

Based on your performance metrics, answer the following questions:

- Explain the impact of accuracy vs. F1 score.
- In this particular problem, which metric one should focus more on?
- Is there a better metric suitable for the use case of vulnerability prediction? Why?


### Answers

The validation dataset contains **1000 samples**, of which **283 are vulnerable** and **717 are non-vulnerable**. With the default threshold of 0.5, the model produces **TP=20, TN=716, FP=1, FN=263**, giving **accuracy=0.736**, **precision=0.952**, **recall=0.071**, and **F1-score=0.132**.

- **Accuracy vs. F1-score:** Accuracy measures the share of all predictions that are correct. Here, the accuracy looks acceptable because the model correctly classifies almost all non-vulnerable samples. However, this hides the fact that the model misses most vulnerable samples. F1-score is more informative because it combines precision and recall for the vulnerable class.
- **Metric to focus on here:** For vulnerability prediction, accuracy is not enough. The recall is very low, which means many real vulnerabilities are classified as non-vulnerable. F1-score is a better single-number summary than accuracy, but recall should be watched very closely.
- **Better metric for this use case:** Recall for the vulnerable class is especially important because missing a real vulnerability is usually more harmful than raising a false alarm. In practice, we can inspect recall together with precision, F1-score, and a precision-recall curve. This makes the decision threshold explicit and lets us choose an operating point that catches more vulnerabilities while keeping the number of false alarms manageable.
